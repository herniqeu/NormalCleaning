```python
from google.colab import drive
import pandas as pd 
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_classification
import numpy as np 
import plotly.express as px

drive.mount('/content/drive')

np.random.seed(73)
```

## Mean

```python
def Mean(numberOfNanX): 
  df_train = pd.read_csv('/content/drive/MyDrive/house-prices-advanced-regression-techniques/train.csv')
  df_test = pd.read_csv('/content/drive/MyDrive/house-prices-advanced-regression-techniques/test.csv')

  df_train = df_train[['LotArea','LotFrontage','SalePrice']]
  df_test = df_test[['LotArea','LotFrontage']]

  df_train.loc[:numberOfNanX, 'LotFrontage'] = np.nan

  LotFrontageMean = df_train['LotFrontage'].mean(skipna=True)
  df_train['LotFrontage'] = df_train['LotFrontage'].fillna(LotFrontageMean)
  df_test['LotFrontage'] = df_test['LotFrontage'].fillna(LotFrontageMean)

  X_train = df_train.drop(columns=['SalePrice'])
  y_train = df_train['SalePrice']

  X_test = df_test

  rfMean = RandomForestRegressor(max_depth=9)
  rfMean.fit(X_train,y_train)

  return rfMean.score(X_train,y_train)
```

# Media

```python
def Median(numberOfNanX): 
  df_train = pd.read_csv('/content/drive/MyDrive/house-prices-advanced-regression-techniques/train.csv')
  df_test = pd.read_csv('/content/drive/MyDrive/house-prices-advanced-regression-techniques/test.csv')

  df_train = df_train[['LotArea','LotFrontage','SalePrice']]
  df_test = df_test[['LotArea','LotFrontage']]

  df_train.loc[:numberOfNanX, 'LotFrontage'] = np.nan

  LotFrontageMedian = df_train['LotFrontage'].median()
  df_train['LotFrontage'] = df_train['LotFrontage'].fillna(LotFrontageMedian)
  df_test['LotFrontage'] = df_test['LotFrontage'].fillna(LotFrontageMedian)


  X_train = df_train.drop(columns=['SalePrice'])
  y_train = df_train['SalePrice']

  X_test = df_test

  rfMedian = RandomForestRegressor(max_depth=9)
  rfMedian.fit(X_train,y_train)

  return rfMedian.score(X_train,y_train)
```

# Mode

```python
def Mode(numberOfNanX): 
  df_train = pd.read_csv('/content/drive/MyDrive/house-prices-advanced-regression-techniques/train.csv')
  df_test = pd.read_csv('/content/drive/MyDrive/house-prices-advanced-regression-techniques/test.csv')

  df_train = df_train[['LotArea','LotFrontage','SalePrice']]
  df_test = df_test[['LotArea','LotFrontage']]

  df_train.loc[:numberOfNanX, 'LotFrontage'] = np.nan

  LotFrontageMode = df_train['LotFrontage'].mode()[0]
  df_train['LotFrontage'] = df_train['LotFrontage'].fillna(LotFrontageMode)
  df_test['LotFrontage'] = df_test['LotFrontage'].fillna(LotFrontageMode)

  X_train = df_train.drop(columns=['SalePrice'])
  y_train = df_train['SalePrice']

  X_test = df_test

  rfMode = RandomForestRegressor(max_depth=9)
  rfMode.fit(X_train,y_train)

  return rfMode.score(X_train,y_train)
```

# Normal Distribution

```python
def NormalDistribuition(numberOfNanX): 
  df_train = pd.read_csv('/content/drive/MyDrive/house-prices-advanced-regression-techniques/train.csv')
  df_test = pd.read_csv('/content/drive/MyDrive/house-prices-advanced-regression-techniques/test.csv')

  df_train = df_train[['LotArea','LotFrontage','SalePrice']]
  df_test = df_test[['LotArea','LotFrontage']]

  df_train.loc[:numberOfNanX, 'LotFrontage'] = np.nan

  LotFrontageMean = df_train['LotFrontage'].mean(skipna=True)
  LotFrontageStd = df_train['LotFrontage'].std(skipna=True)

  num_nulls = df_train['LotFrontage'].isnull().sum()
  num_nulls_test = df_test['LotFrontage'].isnull().sum()

  lista_numeros_normal = np.random.normal(LotFrontageMean, LotFrontageStd, num_nulls)
  lista_numeros_normal_test = np.random.normal(LotFrontageMean, LotFrontageStd, num_nulls)[0: num_nulls_test]

  df_train['LotFrontage'].fillna(pd.Series(lista_numeros_normal, index=df_train.index[df_train['LotFrontage'].isnull()]), inplace=True)

  df_test['LotFrontage'].fillna(pd.Series(lista_numeros_normal_test, index=df_test.index[df_test['LotFrontage'].isnull()]), inplace=True)

  X_train = df_train.drop(columns=['SalePrice'])
  y_train = df_train['SalePrice']

  X_test = df_test

  rfNormalDistribuition = RandomForestRegressor(max_depth=9)
  rfNormalDistribuition.fit(X_train,y_train)

  return rfNormalDistribuition.score(X_train,y_train)
```

```python
TestList = []
for i in range(0, 1201, 50):
  TestList.append(i) 
```

```python
ResultsList = [[],[],[],[]]
for i in TestList:
  ResultsList[0].append(Mean(i))
  ResultsList[1].append(Median(i))
  ResultsList[2].append(Mode(i))
  ResultsList[3].append(NormalDistribuition(i))
```

```python
data = {'Mean': ResultsList[0],
        'Median': ResultsList[1],
        'Mode': ResultsList[2],
        'Normal Distribution': ResultsList[3]
        }

df_Results = pd.DataFrame(data)
df_Results.index = TestList
```
